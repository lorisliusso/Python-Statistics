# Legge dei Grandi Numeri! (Law of Large Numbers): ⚖️

La legge dei grandi numeri è un teorema fondamentale della teoria della probabilità che indica che se ripetiamo molte volte (tendendo all'infinito) lo stesso esperimento, la frequenza di un certo evento tende ad essere costante.

Vale a dire, la legge dei grandi numeri indica che se lo stesso test viene eseguito ripetutamente (ad esempio, lancio di una moneta), la frequenza con cui si ripeterà un determinato evento (testa/croce) si avvicinerà a una costante. Questa a sua volta sarà la probabilità che questo evento si verifichi.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(123)

def heads(sample_size):
    
    '''
    funzione che simula il lancio di una moneta per n. volte (sample_size) 
    e restituisce il numero di teste e la relativa frequenza di uscita.
    '''
    
    trials_list= [np.random.randint(0, 1+1) for i in range(sample_size)] #list of n. tails (0) and n. heads (1)
    
    n_heads= sum(trials_list)
    
    freq= (n_heads/sample_size)*100
    
    print(f'Number of heads out of {sample_size}: {n_heads} -> {freq}%')
    

#Let's try:

heads(10)
heads(100)
heads(1000)
heads(10000)
heads(100000)
```

```python
np.random.seed(123)

#Proviamo a calcolare la frequenza con sample_size di grandezze (numero di lanci) sempre maggiori da 10 a 20000:

sample_size_list= np.arange(10,20000, 100) 
freq_list= [(sum(np.random.randint(0, 1+1) for i in range(size))/size) for size in sample_size_list]
```

```python
#Andiamo a visualizzare i risultati ottenuti:

plt.figure(figsize=(14,8))

plt.axhline(y=0.5, color='orange', linestyle='-', linewidth=5)

plt.plot(sample_size_list, freq_list, color='black', linewidth=2)
plt.title('Average number of heads per coin flip over a sample size')
plt.xlabel('Sample size')
plt.ylabel('Average number of heads')
plt.show()
```
